## 1. Hiểu dữ liệu (Data Understanding)



### ✔ Kiểm tra cơ bản
- Số lượng tuần  
- Missing values  
- Trùng lặp  
- Khoảng thời gian có liên tục theo tuần không  
- Tính seasonality dạng tuần/lễ/tết  



### ✔ Chuyển đổi kiểu dữ liệu
- Convert `time_index` → `datetime`  
- Set `time_index` làm index  
- Kiểm tra tần suất có đúng `W` không  
  - Nếu lệch, cần chuẩn hóa bằng `resample`.



---



## 2. Phân tích khám phá dữ liệu (EDA – Exploratory Data Analysis)



### 2.1 Visual Exploration
- Plot đường doanh thu theo thời gian  
- Plot rolling mean (4 tuần, 12 tuần)  
- Plot seasonal decomposition gồm:  
  - trend  
  - seasonality (theo tuần trong năm, theo tháng)  
  - residual  



### 2.2 Kiểm tra tính stationarity
- Augmented Dickey-Fuller Test  
→ Quyết định có cần differencing trước khi dùng ARIMA/ETS hay không.



### 2.3 Phân tích phân phối & thống kê
- Histogram & boxplot doanh thu  
- Weekly YoY % change  
- Impact của campaign/holiday (nếu có)



---



## 3. Feature Engineering cho Time-Series



### ✔ Tạo các feature



#### Lag features:
- `lag_1`, `lag_2`, `lag_4`, `lag_12` (tùy seasonality)



#### Rolling windows:
- `MA_4` (monthly)  
- `MA_12` (quarterly)



#### Dummies:
- tuần trong năm (1–52)  
- tháng trong năm (1–12)  
- holiday flag  



### ✔ Xử lý đặc thù retail
- Weekly seasonality mạnh theo event (Tết, Black Friday…)  
- Nên thêm feature *special events* nếu có lịch.



---



## 4. Chọn mô hình & Kiểm tra khả năng dự báo
Bạn sẽ test nhiều loại model để tìm ra mô hình phù hợp nhất.



### 4.1 Baseline models
- Naive  
- Seasonal Naive  
- Moving Average  



### 4.2 Models truyền thống
- ARIMA / SARIMA  
- Holt-Winters Exponential Smoothing (ETS)



### 4.3 Models hiện đại
- Facebook Prophet (hợp dữ liệu retail weekly)  



#### Machine Learning:
- Random Forest Regressor  
- XGBoost / LightGBM  



#### Deep Learning:
- LSTM  
- Temporal Fusion Transformer (TFT)  
- N-BEATS  



> Lưu ý: nếu < 100 tuần → ưu tiên SARIMA/ETS/Prophet.



---



## 5. Chuẩn hóa cách đánh giá mô hình (Backtesting)



### ✔ Chia tập theo Time-Series Cross Validation
Ví dụ:  
- Train
- Test
- Rolling windows: forecast 4 tuần mỗi vòng  



### ✔ Metrics quan trọng
- MAPE  
- sMAPE  
- RMSE  
- MAE  
- WAPE / Bias  



Retail forecasting thường ưu tiên: **WAPE + Bias**.



---



## 6. Lựa chọn mô hình tốt nhất
Sau khi backtest:



### Chọn mô hình có:
- WAPE thấp nhất  
- Bias gần 0 (không quá dự đoán hoặc thấp quá)  
- Forecast ổn định, không giật mạnh  



Thông thường cho dữ liệu doanh thu theo tuần:  
**Prophet, SARIMA, hoặc N-BEATS** cho kết quả tốt.
