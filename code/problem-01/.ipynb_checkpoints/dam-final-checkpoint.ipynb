# 1) Import libs, Load data and Calculate TotalPrice

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import STL
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from scipy.stats import boxcox

# ----------------------------
# 1) Load data (robust paths)
# ----------------------------
DATA_PATH_CANDIDATES = [
    "data.csv",
    "/kaggle/input/ecommerce-data/data.csv",
    "/kaggle/input/ecommerce-data/data.csv".replace("data.csv", "data.csv"),
]

data_path = None
for p in DATA_PATH_CANDIDATES:
    try:
        _ = pd.read_csv(p, nrows=5, encoding="latin1")
        data_path = p
        break
    except Exception:
        pass

if data_path is None:
    raise FileNotFoundError(
        "Không tìm thấy file data.csv. Hãy đặt data.csv cùng thư mục notebook hoặc cập nhật DATA_PATH_CANDIDATES."
    )

df = pd.read_csv(data_path, encoding="latin1")

# ----------------------------
# 2) Basic cleaning / features
# ----------------------------
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
df = df.dropna(subset=["InvoiceDate"]).copy()

df["Revenue"] = df["Quantity"] * df["UnitPrice"]

# Chỉ giữ doanh thu dương cho phân tích xu hướng doanh thu (trả hàng/hoàn tiền làm phần riêng nếu cần)
df_pos = df[df["Revenue"] > 0].copy()

print("Loaded rows:", len(df), "| Positive revenue rows:", len(df_pos))
print("Date range:", df_pos["InvoiceDate"].min(), "->", df_pos["InvoiceDate"].max())
```

# 2) Time Normalization

```python
# -----------------------------------------
# YC2 - Time index normalization (DAILY)
# -----------------------------------------
daily_raw = df_pos.set_index("InvoiceDate").resample("D")["Revenue"].sum()

# Create a complete daily index to detect missing timestamps explicitly
full_range = pd.date_range(start=daily_raw.index.min(), end=daily_raw.index.max(), freq="D")
missing_days = full_range.difference(daily_raw.index)

print("Daily points (raw):", len(daily_raw))
print("Missing days (no record after resample):", len(missing_days))
if len(missing_days) > 0:
    print("Sample missing days:", list(missing_days[:10]))

# In online retail, a missing day after daily resample often means 'no transactions that day'
# -> fill 0 at this stage (after we've counted missing timestamps)
daily = daily_raw.reindex(full_range, fill_value=0)
daily.index.name = "Date"

print("\nDaily series summary:")
print(daily.describe())
```

# 3) Observation

```python
plt.figure(figsize=(14,5))
plt.plot(daily)
plt.title("Daily Revenue Time Series")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(daily, label="Daily Revenue", alpha=0.4)
plt.plot(daily.rolling(30).mean(), label="30-day Rolling Mean (Trend)", linewidth=2)
plt.title("Trend Estimation using Rolling Mean")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.show()
```

# 4) Decompose Trend / Seasonality / Noise

```python
# -----------------------------------------
# YC1 - Decompose Trend / Seasonality / Noise
# -----------------------------------------
# Weekly seasonality is common in retail => period=7 for daily data
stl = STL(daily, period=7, robust=True)
stl_res = stl.fit()

trend_stl = stl_res.trend
seasonal_stl = stl_res.seasonal
resid_stl = stl_res.resid

# Plots
fig = stl_res.plot()
fig.set_size_inches(14, 8)
plt.tight_layout()
plt.show()

# -----------------------------------------
# YC1 - Autocorrelation (evidence of seasonality)
# -----------------------------------------
plt.figure(figsize=(12,4))
plot_acf(daily, lags=60, ax=plt.gca())
plt.title("ACF of Daily Revenue (check weekly/monthly patterns)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plot_pacf(daily, lags=60, ax=plt.gca(), method="ywm")
plt.title("PACF of Daily Revenue")
plt.tight_layout()
plt.show()
```

# 8) Outlier Detection

YC2

```python
# -----------------------------------------
# YC2 - Outlier detection / handling (rolling, time-aware)
# -----------------------------------------
# Rolling quantile caps reduce the impact of extreme spikes while keeping general shape.
# Window=30 days is a practical baseline for daily retail series.
window = 30
roll_q05 = daily.rolling(window, center=True, min_periods=7).quantile(0.05)
roll_q95 = daily.rolling(window, center=True, min_periods=7).quantile(0.95)

# For edges (where rolling isn't available), fall back to global quantiles
global_q05 = daily.quantile(0.05)
global_q95 = daily.quantile(0.95)
roll_q05 = roll_q05.fillna(global_q05)
roll_q95 = roll_q95.fillna(global_q95)

is_outlier_rolling = (daily < roll_q05) | (daily > roll_q95)
print("Rolling outliers detected:", int(is_outlier_rolling.sum()))
print("Outlier share:", float(is_outlier_rolling.mean()))

# Cap outliers (winsorize time-aware)
daily_clean = daily.clip(lower=roll_q05, upper=roll_q95)

# Keep a copy of original for event analysis later
daily_original = daily.copy()
```

```python
# (Optional) Global IQR outliers - quick sanity check (not time-aware)
Q1 = daily_original.quantile(0.25)
Q3 = daily_original.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_global = daily_original[(daily_original < lower) | (daily_original > upper)]
print("Detected global IQR outliers:", len(outliers_global))
print(outliers_global.head(10))
```

# 11) Kiểm tra tính dừng

```python
def adf_test(series, name="Series"):
    result = adfuller(series.dropna())
    print(f"ADF Statistic for {name}: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] < 0.05:
        print("=> Chuỗi DỪNG")
    else:
        print("=> Chuỗi KHÔNG dừng")
```

```python
daily_log = np.log1p(daily_winsor)
```

```python
# -----------------------------------------
# YC2 - Stationarity checks (for modeling readiness)
# -----------------------------------------
adf_test(daily_winsor, "Daily Revenue (Cleaned)")
adf_test(daily_log, "Log1p Transformed (Cleaned)")

daily_diff = daily_log.diff().dropna()
adf_test(daily_diff, "Differenced Log1p Series")
```

# 12) Save files

```python
# =========================
# YC3 - Time Series Decomposition (Trend / Seasonality / Residual)
# + Save outputs to files (CSV/XLSX/PNG)
# =========================

import os
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.seasonal import STL, seasonal_decompose

# -------------------------
# 0) Output folder + file prefix
# -------------------------
OUT_DIR = "outputs_yc3"
os.makedirs(OUT_DIR, exist_ok=True)

run_id = datetime.now().strftime("%Y%m%d_%H%M%S")
file_prefix = f"yc3_decomposition_{run_id}"

# -------------------------
# 1) Choose the best available daily series (cleaned if exists)
# -------------------------
series_candidates = []
for name in ["daily_clean", "daily", "daily_winsor", "daily_raw"]:
    if name in globals():
        s = globals()[name]
        if isinstance(s, (pd.Series, pd.DataFrame)):
            series_candidates.append((name, s))

if len(series_candidates) == 0:
    raise ValueError("No daily time series found. Please run YC1 & YC2 preprocessing cells first.")

# Prefer cleaned series if available (instead of just taking first found)
preferred_order = ["daily_clean", "daily_winsor", "daily", "daily_raw"]
series_candidates = sorted(series_candidates, key=lambda x: preferred_order.index(x[0]) if x[0] in preferred_order else 999)

chosen_name, chosen_obj = series_candidates[0]

# Convert to 1D Series
if isinstance(chosen_obj, pd.DataFrame):
    ts = chosen_obj.iloc[:, 0].copy()
else:
    ts = chosen_obj.copy()

ts = ts.sort_index()
ts = ts.asfreq("D")

# Safety: fill any remaining missing values (STL cannot handle NaN)
# Interpolate by time first, then fill remaining gaps with 0 (rare)
ts_filled = ts.interpolate(method="time").fillna(0)

print(f"Using series: {chosen_name}")
print(f"Start: {ts_filled.index.min()} | End: {ts_filled.index.max()} | Points: {len(ts_filled)}")
print(f"Missing after asfreq (before fill): {ts.isna().sum()}")

# -------------------------
# 2) STL Decomposition (recommended)
# -------------------------
PERIOD = 7  # weekly seasonality for daily retail data

stl = STL(ts_filled, period=PERIOD, robust=True)
stl_result = stl.fit()

trend = stl_result.trend
seasonal = stl_result.seasonal
resid = stl_result.resid

# Plot STL decomposition and SAVE to PNG
fig_stl = stl_result.plot()
fig_stl.set_size_inches(12, 8)
plt.suptitle(f"STL Decomposition (period={PERIOD}, robust=True) | series={chosen_name}", y=1.02)
plt.tight_layout()

stl_png_path = os.path.join(OUT_DIR, f"{file_prefix}_stl.png")
fig_stl.savefig(stl_png_path, dpi=200, bbox_inches="tight")
plt.show()
plt.close(fig_stl)

# -------------------------
# 3) Classical decomposition (optional) + SAVE
# -------------------------
classical_png_path = None
try:
    decomp_add = seasonal_decompose(ts_filled, model="additive", period=PERIOD, extrapolate_trend="freq")
    fig_classical = decomp_add.plot()
    fig_classical.set_size_inches(12, 8)
    plt.suptitle(f"Classical Decomposition - Additive (period={PERIOD}) | series={chosen_name}", y=1.02)
    plt.tight_layout()

    classical_png_path = os.path.join(OUT_DIR, f"{file_prefix}_classical_additive.png")
    fig_classical.savefig(classical_png_path, dpi=200, bbox_inches="tight")
    plt.show()
    plt.close(fig_classical)

except Exception as e:
    print("Classical additive decomposition skipped due to:", str(e))

# -------------------------
# 4) Quick sanity metrics
# -------------------------
print("\n--- Residual sanity check ---")
print("Residual mean:", float(np.nanmean(resid)))
print("Residual std:", float(np.nanstd(resid)))

# -------------------------
# 5) Save components for later steps (CSV + Excel)
# -------------------------
decomposition_components = pd.DataFrame({
    "observed": ts_filled,
    "trend": trend,
    "seasonal": seasonal,
    "residual": resid
})

csv_path = os.path.join(OUT_DIR, f"{file_prefix}_components.csv")
xlsx_path = os.path.join(OUT_DIR, f"{file_prefix}_components.xlsx")

decomposition_components.to_csv(csv_path, index=True)
decomposition_components.to_excel(xlsx_path, index=True)

print("\n--- Saved files ---")
print("STL plot PNG:", stl_png_path)
if classical_png_path:
    print("Classical plot PNG:", classical_png_path)
print("Components CSV:", csv_path)
print("Components Excel:", xlsx_path)

decomposition_components.head()
```
