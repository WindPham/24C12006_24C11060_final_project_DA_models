```python
import os

# Nội dung mẫu cho file HTML
template = '''<!DOCTYPE html>
<html lang="vi">
<head>
    <meta charset="utf-8">
    <style>
        body {{
            background-color: transparent; /* ăn nền từ index.html */
            color: #0f172a; /* slate-900 */
            font-family: 'Segoe UI', sans-serif;
            margin: 0;
            height: 100vh;
            overflow: hidden;
            display: flex;
            flex-direction: column;
            justify-content: center;
            padding: 60px 80px;
            box-sizing: border-box;
        }}

        /* --- ĐỀ MỤC GÓC TRÁI TRÊN --- */
        .section-title {{
            position: absolute;
            top: 28px;
            left: 40px;
            font-family: Arial, sans-serif;
            font-weight: 650;
            font-size: 36px;
            letter-spacing: 2px;
            text-transform: uppercase;
            color: rgba(15, 23, 42, 0.55);
        }}

        h1 {{
            color: #0077b6;
            font-size: 56px;
            margin: 0 0 20px 0;
        }}

        .content {{
            font-size: 28px;
            line-height: 1.5;
            border-left: 4px solid #0077b6;
            padding-left: 22px;
            color: #1e293b; /* slate-800 */
            max-width: 900px;
        }}
    </style>
</head>
<body>

    <div class="section-title">Giới thiệu</div>

    <h1>SLIDE {slide_number:02d}</h1>

    <div class="content">
        Định dạng layout khoa học dữ liệu của bạn ở đây.
    </div>

</body>
</html>
'''

# Tạo thư mục slides nếu chưa tồn tại
slides_dir = "D:/career/master/da_models/final_project/presentation/slides"
if not os.path.exists(slides_dir):
    os.makedirs(slides_dir)
    print(f"Đã tạo thư mục: {slides_dir}")

# Số lượng slide muốn tạo
num_slides = 20  # Thay đổi số này nếu muốn tạo nhiều hơn ít hơn slide

# Tạo các file slide
for i in range(1, num_slides + 1):
    filename = f"slide-{i}.html"  # Định dạng số với 2 chữ số trong tên file
    filepath = os.path.join(slides_dir, filename)
    
    # Tạo nội dung với số slide đã định dạng
    content = template.format(slide_number=i)
    
    # Ghi file
    with open(filepath, 'w', encoding='utf-8') as file:
        file.write(content)
    
    print(f"Đã tạo: {filename}")

print(f"\nĐã tạo thành công {num_slides} slide trong thư mục '{slides_dir}'")
```
