```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# 0) READ DATA
# =========================
# Nu bn chy notebook cng th mc vi file:
TRANSACTION_PATH = "/kaggle/input/damlogdataset/onlret_transaction.csv"
DESC_PATH = "/kaggle/input/damlogdataset/onlret_description.csv"  # optional

df = pd.read_csv(TRANSACTION_PATH)

print("Raw shape:", df.shape)
print(df.head(3))

# =========================
# 1) BASIC CLEANING
# =========================

# 1.1 Chun ha tn ct nu dataset c khc nhau (bn c th sa li cho ng ct thc t)
# Cc ct thng gp trong Online Retail:
# InvoiceNo, StockCode, Description, Quantity, InvoiceDate, UnitPrice, CustomerID, Country

# 1.2 Parse datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# 1.3 Convert numeric
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"], errors="coerce")

# 1.4 Drop rows missing key fields
df = df.dropna(subset=["InvoiceDate", "Quantity", "UnitPrice"]).copy()

# 1.5 Optional: remove weird extreme values (tu dataset)
# df = df[df["UnitPrice"] >= 0]

print("\nAfter basic cleaning shape:", df.shape)

# =========================
# 2) CREATE REVENUE
# =========================
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

# Thm ct ngy (daily key)
df["date"] = df["InvoiceDate"].dt.floor("D")

# =========================
# 3) OPTIONAL: HANDLE CANCEL INVOICES (if exists)
# =========================
# Nhiu dataset c Cancel invoice bt u bng "C"
# Nu ct InvoiceNo c trong d liu -> bn c th chn lc cancel invoice
if "InvoiceNo" in df.columns:
    df["InvoiceNo"] = df["InvoiceNo"].astype(str)
    cancel_mask = df["InvoiceNo"].str.startswith("C")

    print("\nCancel invoices count:", cancel_mask.sum())

    # Nu bn mun LOI cancel invoice khi phn tch:
    # df = df[~cancel_mask].copy()
    #
    # Nu bn mun GI returns  phn tch (gross/returns/net) th KHNG drop  y.

# =========================
# 4) BUILD DAILY GROSS / RETURNS / NET
# =========================

# 4.1 Gross revenue: ch ly revenue dng
daily_gross = df.loc[df["Revenue"] > 0].groupby("date")["Revenue"].sum()

# 4.2 Returns revenue: revenue m (tr hng), ly gi tr tuyt i  d nhn
daily_returns = df.loc[df["Revenue"] < 0].groupby("date")["Revenue"].sum()
daily_returns_abs = daily_returns.abs()

# 4.3 Net revenue = gross - returns
daily_net = daily_gross.sub(daily_returns_abs, fill_value=0)

# 4.4 Align full daily date range
full_range = pd.date_range(
    start=df["date"].min(),
    end=df["date"].max(),
    freq="D"
)

daily_gross = daily_gross.reindex(full_range, fill_value=0)
daily_returns_abs = daily_returns_abs.reindex(full_range, fill_value=0)
daily_net = daily_net.reindex(full_range, fill_value=0)

# 4.5 Return rate = returns / gross (trnh chia 0)
daily_return_rate = daily_returns_abs / daily_gross.replace(0, np.nan)
daily_return_rate = daily_return_rate.fillna(0)

# 4.6 Tng hp bng kt qu daily
daily_revenue_summary = pd.DataFrame({
    "gross_revenue": daily_gross,
    "returns_value": daily_returns_abs,
    "net_revenue": daily_net,
    "return_rate": daily_return_rate
})

print("\nDaily summary shape:", daily_revenue_summary.shape)
print(daily_revenue_summary.head())

# =========================
# 5) QUICK CHECKS (QA)
# =========================
print("\n--- Quick stats ---")
print("Total gross revenue:", daily_revenue_summary["gross_revenue"].sum())
print("Total returns value:", daily_revenue_summary["returns_value"].sum())
print("Total net revenue:", daily_revenue_summary["net_revenue"].sum())
print("Avg return rate (daily):", daily_revenue_summary["return_rate"].mean())

# =========================
# 6) PLOTS (for YC1 EDA)
# =========================

# Plot 1: Gross vs Returns vs Net
plt.figure(figsize=(14, 5))
plt.plot(daily_revenue_summary.index, daily_revenue_summary["gross_revenue"], label="Gross Revenue (Sales)")
plt.plot(daily_revenue_summary.index, daily_revenue_summary["returns_value"], label="Returns (Abs Value)")
plt.plot(daily_revenue_summary.index, daily_revenue_summary["net_revenue"], label="Net Revenue")
plt.title("Daily Revenue Overview: Gross vs Returns vs Net")
plt.legend()
plt.tight_layout()
plt.show()

# Plot 2: Return Rate
plt.figure(figsize=(14, 4))
plt.plot(daily_revenue_summary.index, daily_revenue_summary["return_rate"], label="Return Rate")
plt.title("Daily Return Rate = Returns / Gross")
plt.legend()
plt.tight_layout()
plt.show()

# =========================
# 7) SAVE OUTPUT FILES (optional)
# =========================
daily_revenue_summary.to_csv("daily_revenue_summary_gross_returns_net.csv", index=True)
print("\nSaved file: daily_revenue_summary_gross_returns_net.csv")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.seasonal import STL, seasonal_decompose

# =========================
# 0) READ DATA (2 files)
# =========================
TRANSACTION_PATH = "/kaggle/input/damdataset/onlret_transaction.csv"
DESC_PATH = "/kaggle/input/damdataset/onlret_description.csv"

df_txn = pd.read_csv(TRANSACTION_PATH)
df_desc = pd.read_csv(DESC_PATH)

print(" Transaction shape:", df_txn.shape)
print(" Desc shape:", df_desc.shape)
print("Transaction columns:", df_txn.columns.tolist())
print("Desc columns:", df_desc.columns.tolist())

# =========================
# 1) AUTO DETECT COLUMNS
# =========================
def find_col(possible_names, columns):
    cols_lower = {c.lower(): c for c in columns}
    for name in possible_names:
        if name.lower() in cols_lower:
            return cols_lower[name.lower()]
    return None

date_col = find_col(["InvoiceDate", "invoice_date", "Date", "date", "Datetime", "datetime"], df_txn.columns)
qty_col  = find_col(["Quantity", "quantity", "Qty", "qty"], df_txn.columns)
price_col= find_col(["UnitPrice", "unitprice", "Unit Price", "unit_price", "Price", "price"], df_txn.columns)
inv_col  = find_col(["InvoiceNo", "invoice_no", "Invoice", "invoice"], df_txn.columns)

print("\n--- Auto-detected columns ---")
print("date_col  =", date_col)
print("qty_col   =", qty_col)
print("price_col =", price_col)
print("inv_col   =", inv_col)

if date_col is None or qty_col is None or price_col is None:
    raise ValueError(" Missing required columns for time series analysis (date/quantity/unit price).")

# =========================
# 2) BASIC CLEANING + REVENUE
# =========================
df = df_txn.copy()

df[date_col]  = pd.to_datetime(df[date_col], errors="coerce")
df[qty_col]   = pd.to_numeric(df[qty_col], errors="coerce")
df[price_col] = pd.to_numeric(df[price_col], errors="coerce")

df = df.dropna(subset=[date_col, qty_col, price_col]).copy()
df["date"] = df[date_col].dt.floor("D")

# Revenue can be negative because of returns (Quantity < 0)
df["Revenue"] = df[qty_col] * df[price_col]

# Optional: attach description for later deep dive
# join by StockCode if exists
if "StockCode" in df.columns and "StockCode" in df_desc.columns:
    df = df.merge(df_desc.drop_duplicates(subset=["StockCode"]), on="StockCode", how="left")

print("\n After cleaning shape:", df.shape)
print("Revenue min/max:", df["Revenue"].min(), df["Revenue"].max())

# ============================================================
# YC1 (SMART): EDA daily/weekly/monthly + trend/seasonality/noise
# ============================================================

# =========================
# YC1.1 Build daily Gross / Returns / Net
# =========================
daily_gross = df.loc[df["Revenue"] > 0].groupby("date")["Revenue"].sum()
daily_returns = df.loc[df["Revenue"] < 0].groupby("date")["Revenue"].sum().abs()
daily_net = daily_gross.sub(daily_returns, fill_value=0)

full_range = pd.date_range(df["date"].min(), df["date"].max(), freq="D")
daily_gross = daily_gross.reindex(full_range, fill_value=0)
daily_returns = daily_returns.reindex(full_range, fill_value=0)
daily_net = daily_net.reindex(full_range, fill_value=0)

daily_summary = pd.DataFrame({
    "gross_revenue": daily_gross,
    "returns_value": daily_returns,
    "net_revenue": daily_net
})
daily_summary["return_rate"] = daily_summary["returns_value"] / daily_summary["gross_revenue"].replace(0, np.nan)
daily_summary["return_rate"] = daily_summary["return_rate"].fillna(0)

print("\n Daily summary built:", daily_summary.shape)
print(daily_summary.head())

# =========================
# YC1.2 SMART - Daily / Weekly / Monthly revenue
# =========================
weekly_summary = daily_summary.resample("W")["gross_revenue"].sum().to_frame("gross_revenue_weekly")
monthly_summary = daily_summary.resample("M")["gross_revenue"].sum().to_frame("gross_revenue_monthly")

# =========================
# YC1.3 SMART - Rolling mean + Coefficient of Variation (CV)
# =========================
# Rolling mean 7 days & 30 days
daily_summary["rm_7"] = daily_summary["gross_revenue"].rolling(7).mean()
daily_summary["rm_30"] = daily_summary["gross_revenue"].rolling(30).mean()

# CV = std/mean (rolling)
daily_summary["cv_7"] = daily_summary["gross_revenue"].rolling(7).std() / daily_summary["gross_revenue"].rolling(7).mean()
daily_summary["cv_30"] = daily_summary["gross_revenue"].rolling(30).std() / daily_summary["gross_revenue"].rolling(30).mean()

# =========================
# YC1.4 SMART - WoW / MoM growth (Measurable)
# =========================
weekly_summary["WoW_growth"] = weekly_summary["gross_revenue_weekly"].pct_change()
monthly_summary["MoM_growth"] = monthly_summary["gross_revenue_monthly"].pct_change()

# =========================
# YC1.5 PLOTS (Daily + smoothing)
# =========================
plt.figure(figsize=(14,5))
plt.plot(daily_summary.index, daily_summary["gross_revenue"], label="Daily Gross Revenue")
plt.plot(daily_summary.index, daily_summary["rm_7"], label="Rolling Mean 7D")
plt.plot(daily_summary.index, daily_summary["rm_30"], label="Rolling Mean 30D")
plt.title("YC1 - Daily Gross Revenue with Smoothing (7D & 30D)")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(14,4))
plt.plot(daily_summary.index, daily_summary["cv_7"], label="CV 7D")
plt.plot(daily_summary.index, daily_summary["cv_30"], label="CV 30D")
plt.title("YC1 - Rolling Coefficient of Variation (Volatility)")
plt.legend()
plt.tight_layout()
plt.show()

# Weekly WoW growth
plt.figure(figsize=(14,4))
plt.plot(weekly_summary.index, weekly_summary["WoW_growth"], label="WoW Growth")
plt.axhline(0, linewidth=1)
plt.title("YC1 - Weekly WoW Growth (Gross Revenue)")
plt.legend()
plt.tight_layout()
plt.show()

# Monthly MoM growth
plt.figure(figsize=(14,4))
plt.plot(monthly_summary.index, monthly_summary["MoM_growth"], label="MoM Growth")
plt.axhline(0, linewidth=1)
plt.title("YC1 - Monthly MoM Growth (Gross Revenue)")
plt.legend()
plt.tight_layout()
plt.show()

# Monthly deep dive table (Time-bound: focus by month)
monthly_table = monthly_summary.copy()
monthly_table["MoM_growth_%"] = monthly_table["MoM_growth"] * 100
print("\n Monthly deep-dive:")
print(monthly_table)

# ============================================================
# YC2 (SMART): Preprocessing for time series (daily)
# ============================================================

# =========================
# YC2.1 Choose modeling series (Gross is best for demand rhythm)
# =========================
ts = daily_summary["gross_revenue"].copy()
ts = ts.sort_index().asfreq("D")

# Missing timestamp check
missing_days = ts.isna().sum()
print("\n Missing days after asfreq:", missing_days)

# Fill missing carefully
# In retail: missing day often means no sales -> 0, but can be data gap.
ts_filled = ts.fillna(0)

# =========================
# YC2.2 Outlier detection (rolling quantile cap)
# =========================
window = 30
q_low = ts_filled.rolling(window).quantile(0.05)
q_high = ts_filled.rolling(window).quantile(0.95)

# fallback for first window
q_low = q_low.fillna(ts_filled.expanding().quantile(0.05))
q_high = q_high.fillna(ts_filled.expanding().quantile(0.95))

ts_winsor = ts_filled.clip(lower=q_low, upper=q_high)

outlier_flag = (ts_filled < q_low) | (ts_filled > q_high)
print(" Outlier days detected:", int(outlier_flag.sum()))

# =========================
# YC2.3 Transform (log1p) + Stationarity check (optional)
# =========================
ts_log = np.log1p(ts_winsor)
ts_log_diff = ts_log.diff().dropna()

# Save to globals for later use
daily_clean = ts_winsor.copy()
daily_log = ts_log.copy()
daily_log_diff = ts_log_diff.copy()

# ============================================================
# YC3 (SMART): Decomposition for 12 months (Trend/Seasonality/Noise)
# ============================================================

PERIOD = 7  # weekly rhythm for daily retail

# =========================
# LOG TRANSFORM (Recommended for retail revenue)
# - Retail revenue often has variance that increases with level.
# - Using log1p makes the series closer to an additive structure on log-scale:
#   log(1+y) = trend + season + residual  ( multiplicative on original scale)
# =========================
daily_clean_log = np.log1p(daily_clean)

# STL Decomposition (robust) on log-scale
stl = STL(daily_clean_log, period=PERIOD, robust=True)
stl_result = stl.fit()

fig = stl_result.plot()
fig.set_size_inches(12, 8)
plt.suptitle(f"YC3 - STL Decomposition on log1p(Revenue) (period={PERIOD}, robust=True)", y=1.02)
plt.tight_layout()
plt.show()

# Classical decomposition (additive) on log-scale (optional reference)
try:
    decomp_add = seasonal_decompose(daily_clean_log, model="additive", period=PERIOD, extrapolate_trend="freq")
    fig2 = decomp_add.plot()
    fig2.set_size_inches(12, 8)
    plt.suptitle(f"YC3 - Classical Decomposition on log1p(Revenue) (Additive, period={PERIOD})", y=1.02)
    plt.tight_layout()
    plt.show()
except Exception as e:
    print("Classical decomposition skipped:", str(e))

# =========================
# Decomposition tables
# 1) Log-scale components (additive)
# 2) Interpretable original-scale components:
#    - trend_level: expm1(trend_log)  (baseline revenue level)
#    - seasonal_factor: exp(seasonal_log)  (multiplicative factor vs baseline)
#    Note: residual remains on log-scale (shocks/events).
# =========================
decomposition_components_log = pd.DataFrame({
    "observed_log1p": daily_clean_log,
    "trend_log": stl_result.trend,
    "seasonal_log": stl_result.seasonal,
    "residual_log": stl_result.resid
})

decomposition_components = pd.DataFrame({
    "observed": daily_clean,
    "trend_level": np.expm1(stl_result.trend),
    "seasonal_factor": np.exp(stl_result.seasonal),
    "residual_log": stl_result.resid
})

print("\n Decomposition components (log-scale) sample:")
print(decomposition_components_log.head())

print("\n Decomposition components (interpretable original-scale) sample:")
print(decomposition_components.head())
# =========================
# SAVE OUTPUTS (optional)
# =========================
daily_summary.to_csv("daily_summary_gross_returns_net.csv", index=True)
weekly_summary.to_csv("weekly_summary_wow.csv", index=True)
monthly_summary.to_csv("monthly_summary_mom.csv", index=True)
decomposition_components.to_csv("decomposition_components_stl.csv", index=True)

print("\n Saved files:")
print("- daily_summary_gross_returns_net.csv")
print("- weekly_summary_wow.csv")
print("- monthly_summary_mom.csv")
print("- decomposition_components_stl.csv")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# =========================
# 0) (OPTIONAL) READ + BUILD df, date, Revenue
# =========================
# Nu bn  c df sn th b block ny.
# TRANSACTION_PATH = "onlret_transaction.csv"
# df = pd.read_csv(TRANSACTION_PATH)
# # t map ct theo dataset ca bn nu cn:
# date_col, qty_col, price_col = "InvoiceDate", "Quantity", "UnitPrice"
# df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
# df[qty_col] = pd.to_numeric(df[qty_col], errors="coerce")
# df[price_col] = pd.to_numeric(df[price_col], errors="coerce")
# df = df.dropna(subset=[date_col, qty_col, price_col]).copy()
# df["date"] = df[date_col].dt.floor("D")
# df["Revenue"] = df[qty_col] * df[price_col]

# =========================
# 1) Build DAILY series for forecasting (SMART goal: rhythm + trend)
#    Recommend: forecast Gross Revenue (sales only)
# =========================
daily_gross = df.loc[df["Revenue"] > 0].groupby("date")["Revenue"].sum()

full_range = pd.date_range(df["date"].min(), df["date"].max(), freq="D")
y = daily_gross.reindex(full_range, fill_value=0).astype(float)

# Optional smoothing/outlier handling (use if you already created daily_clean)
if "daily_clean" in globals() and isinstance(daily_clean, pd.Series):
    y = daily_clean.reindex(full_range, fill_value=0).astype(float)

# Optional transform to stabilize variance
USE_LOG1P = True
y_model = np.log1p(y) if USE_LOG1P else y

# =========================
# 2) Train/Test split (time-based)
# =========================
test_ratio = 0.2
test_size = int(len(y_model) * test_ratio)

train = y_model.iloc[:-test_size]
test  = y_model.iloc[-test_size:]

print("Train points:", len(train), "Test points:", len(test))
print("Train range:", train.index.min(), "->", train.index.max())
print("Test range:", test.index.min(), "->", test.index.max())

# =========================
# 3) Define 2 SARIMA variants (SMART: weekly vs monthly-ish rhythm)
# =========================
# Variant 1: Weekly rhythm (s=7) - stable, interpretable
cfg_weekly = {
    "name": "SARIMA_weekly_(0,1,1)x(0,1,1,7)",
    "order": (0, 1, 1),
    "seasonal_order": (0, 1, 1, 7)
}

# Variant 2: Monthly-ish rhythm (s30) - supports month-level dynamics
cfg_monthly = {
    "name": "SARIMA_monthly_(1,1,1)x(0,1,1,30)",
    "order": (1, 1, 1),
    "seasonal_order": (0, 1, 1, 30)
}

configs = [cfg_weekly, cfg_monthly]

def fit_forecast_sarima(train_series, steps, order, seasonal_order):
    model = SARIMAX(
        train_series,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res = model.fit(disp=False)
    fc = res.get_forecast(steps=steps).predicted_mean
    return res, fc

def inv_transform(s):
    return np.expm1(s) if USE_LOG1P else s

# =========================
# 4) Fit + Forecast + Evaluate
# =========================
rows = []
preds = {}

for cfg in configs:
    res, fc = fit_forecast_sarima(
        train_series=train,
        steps=len(test),
        order=cfg["order"],
        seasonal_order=cfg["seasonal_order"]
    )
    preds[cfg["name"]] = fc

    # Evaluate in original scale
    y_true = inv_transform(test)
    y_pred = inv_transform(fc)

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = (np.abs((y_true - y_pred) / np.where(y_true == 0, np.nan, y_true))).mean() * 100
    mape = float(np.nan_to_num(mape, nan=0.0))

    rows.append({
        "model": cfg["name"],
        "AIC": res.aic,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE_%": mape
    })

results_df = pd.DataFrame(rows).sort_values(["MAE", "RMSE"])
results_df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
# Gi nh bn  c daily_clean (gross revenue theo ngy,  x l outlier)
y = daily_clean.copy()

# Log transform (an ton vi gi tr 0)
y_log = np.log1p(y)

print("Original scale:")
print(y.describe())

print("\nLog scale:")
print(y_log.describe())
test_ratio = 0.2
test_size = int(len(y_log) * test_ratio)

train_log = y_log.iloc[:-test_size]
test_log  = y_log.iloc[-test_size:]

print("Train:", train_log.index.min(), "", train_log.index.max())
print("Test :", test_log.index.min(), "", test_log.index.max())
model_weekly = SARIMAX(
    train_log,
    order=(0,1,1),
    seasonal_order=(0,1,1,7),
    enforce_stationarity=False,
    enforce_invertibility=False
)
res_weekly = model_weekly.fit(disp=False)
model_monthly = SARIMAX(
    train_log,
    order=(1,1,1),
    seasonal_order=(0,1,1,30),
    enforce_stationarity=False,
    enforce_invertibility=False
)
res_monthly = model_monthly.fit(disp=False)
h = len(test_log)

fc_weekly_log  = res_weekly.get_forecast(steps=h).predicted_mean
fc_monthly_log = res_monthly.get_forecast(steps=h).predicted_mean

# Back to original scale
y_test = np.expm1(test_log)
fc_weekly  = np.expm1(fc_weekly_log)
fc_monthly = np.expm1(fc_monthly_log)
def evaluate(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(
        np.abs((y_true - y_pred) / np.where(y_true == 0, np.nan, y_true))
    ) * 100
    return mae, rmse, np.nan_to_num(mape)
mae_w, rmse_w, mape_w = evaluate(y_test, fc_weekly)
mae_m, rmse_m, mape_m = evaluate(y_test, fc_monthly)

results = pd.DataFrame([
    {
        "Model": "SARIMA_weekly_log",
        "AIC": res_weekly.aic,
        "MAE": mae_w,
        "RMSE": rmse_w,
        "MAPE_%": mape_w
    },
    {
        "Model": "SARIMA_monthly_log",
        "AIC": res_monthly.aic,
        "MAE": mae_m,
        "RMSE": rmse_m,
        "MAPE_%": mape_m
    }
])

results
```
